```python
import pandas as pd

# Get unique country ids
unique_countries = df_conflictology['country_id'].unique()

# Initialize a new DataFrame to hold the results
df_new = pd.DataFrame()

for country in unique_countries:
    # Get data for this country
    df_country = df_conflictology[df_conflictology['country_id'] == country].sort_values(
        'month_id')

    # For each step_pred_ column
    for i in range(1, 37):
        # Create a rolling window of the last 12 ln_ged_sb values, shifted by i months
        df_country[f'step_pred_{i}'] = df_country['ln_ged_sb'].shift(
            i).rolling(12).apply(lambda x: x[-1] if isinstance(x, list) else x)

    # Append this country's data to the new DataFrame
    df_new = pd.concat([df_new, df_country])

# Replace the original DataFrame with the new one
df_conflictology = df_new
```

```python
for step in range(1,37):
    df_conflictology[f'step_pred_{step}'] = df_conflictology.groupby('country_id')['ln_ged_sb'].apply(lambda x: x.shift(step).rolling(12).mean()).tolist()
```

```python
unit = 'country_id'

grouped = df_conflictology.groupby(unit)['ln_ged_sb'].apply(list).reset_index()
df_conflictology['outcome'] = None
outcome_dict = dict(zip(grouped[unit], grouped['ln_ged_sb']))
df_conflictology['outcome'] = df_conflictology[unit].map(outcome_dict)
```
